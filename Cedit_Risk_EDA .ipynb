> As we can observe there are lots of outliers and also there is one huge outlier which is the max value of this particular column and far far away from the other observation which is 117000 thousand


> We can also find out the outlier values and their percentile using IQR method

> because we have only outliers that are greater than upper_bound

```python
Outliers = []

def find_outliers(df_column):
    Q1 = df_column.quantile(0.25)
    Q3 = df_column.quantile(0.75)
    IQR = Q3 - Q1
    Upper_bound = Q3 + 1.5 * IQR

    for i in df_column:
      if i > Upper_bound:
        Outliers.append(i)

    return Outliers
```

```python
Outliers = find_outliers(df["AMT_INCOME_TOTAL"])
len(Outliers)
```

```python
Outliers_perc = (len(Outliers)/len(df["AMT_INCOME_TOTAL"]))*100
Outliers_perc
```

> Reolacing Outliers with 95th percentile value

```python
ninety_fifth_percentile = np.percentile(df["AMT_INCOME_TOTAL"], 95)
df["AMT_INCOME_TOTAL"] = np.where(df["AMT_INCOME_TOTAL"] > ninety_fifth_percentile, ninety_fifth_percentile, df["AMT_INCOME_TOTAL"])
```

```python
plt.figure(figsize=(6,5))

sns.boxplot(data = df, x = "AMT_INCOME_TOTAL", color='skyblue', fliersize=5)
plt.title("Distribution of AMT_INCOME_TOTAL")
plt.xlabel("AMT_INCOME_TOTAL")
plt.show()
```

```python
# Defininf the plotting function
def plotting(df, df0, df1, column):


    fig = plt.figure(figsize=(13,10))

    ax1 = plt.subplot(221)
    # Plotting count plot by category with hue as TARGET
    sns.countplot(x=column, hue='TARGET', data=df, ax=ax1)
    plt.xticks(rotation=90)
    plt.title('Plotting data for target in terms of total count')

    ax2 = plt.subplot(222)
    # ploting data for percentage distribution by target variable
    data = pd.DataFrame()
    data['0'] = ((df0[column].value_counts())/len(df0))
    data['1'] = ((df1[column].value_counts())/len(df1))
    data.plot.bar(ax=ax2)
    plt.title('Plotting data for target in terms of percentage')


    fig.tight_layout()
    plt.show()
```

```python
def plot_grouped_bar(df, column1, column2):

    # Create a cross-tabulation of the two categorical variables
    ct = pd.crosstab(df[column1], df[column2])

    # Plot grouped bar plot
    ct.plot(kind='bar')
    plt.title(f'{column2} by {column1}')
    plt.xlabel(column1)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    plt.legend(title=column2)
    plt.show()
```
